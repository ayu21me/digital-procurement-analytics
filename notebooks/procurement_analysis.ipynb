```python
features = df[[
    "Order Item Quantity",
    "Order Item Product Price",
    "Order Item Total",
    "delivery_delay"
]].dropna()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.01, random_state=42)
features["anomaly"] = model.fit_predict(X_scaled)
```

```python
df["order_month_year"] = df["order date (DateOrders)"].dt.to_period("M")
```

```python
monthly_demand = df.groupby("order_month_year")["Order Item Total"].sum().reset_index()
monthly_demand
```

```python
monthly_demand["order_month_year"] = monthly_demand["order_month_year"].astype(str)
monthly_demand["time_index"] = range(len(monthly_demand))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = monthly_demand[["time_index"]]
y = monthly_demand["Order Item Total"]
```

```python
model_lr = LinearRegression()
model_lr.fit(X, y)
```

```python
monthly_demand["predicted_spend"] = model_lr.predict(X)
```

```python
monthly_demand.head()
```

```python
future_months = pd.DataFrame({
    "time_index": range(len(monthly_demand), len(monthly_demand) + 3)
})

future_predictions = model_lr.predict(future_months)
future_predictions
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_demand["time_index"], monthly_demand["Order Item Total"])
plt.plot(monthly_demand["time_index"], monthly_demand["predicted_spend"])
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y, monthly_demand["predicted_spend"])
mae = mean_absolute_error(y, monthly_demand["predicted_spend"])

r2, mae
```
